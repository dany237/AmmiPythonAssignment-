# *Introduction*
Matplotlib is an amazing visualization library in Python for 2D plots of arrays.
* It is built on NumPy arrays
* It is mostly written in python, a few segments are written in C, Objective-C and Javascript for Platform compatibility.
* Plots help to understand trends, and patterns, and to make correlations.


Some key features:
* Compatibility: Works across various operating systems and supports multiple output formats, such as PNG and PDF.
* Extensibility: Highly customizable, enabling users to customize almost every aspect of a plot, from colors and line styles to annotations and axis properties.
* Plot Types: Supports a wide range of plot types, including line plots, scatter plots, bar charts, histograms and pie charts.
* Integration with NumPy: Allowing users to easily plot data stored in NumPy arrays.

If we do not specify the points on the x-axis, they will get the default values 0, 1, 2, 3 etc., depending on the length of the y-points.

#Markers
You can use the keyword argument marker to emphasize each point with a specified marker:

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o') #look up marker references, you can also try '*', '+', 'H' etc
plt.show()
```

You can use the keyword argument markersize or the shorter version, ms to set the size of the markers.

You can use the keyword argument markeredgecolor or the shorter mec to set the color of the edge of the markers.

You can use the keyword argument markerfacecolor or the shorter mfc to set the color inside the edge of the markers.

You can also use Hexadecimal color values or any of the supported colors

```python
plt.plot(ypoints, marker = 'o', ms = 20, mec = 'red', mfc = '#4CAF50')
# List of colors and their format is avalible in https://matplotlib.org/stable/tutorials/colors/colors.html
```

#Multiple plots in one figure:
The subplots() function takes three arguments that describes the layout of the figure.
The layout is organized in rows and columns, which are represented by the first and second argument.
The third argument represents the index of the current plot.

```python
# Plot 1:
x = np.arange(-2, 5, 0.1)
cos_x = np.cos(x)

plt.subplot(1, 2, 1)  # The figure has 1 row, 2 columns, and this plot is the first plot.
plt.plot(x, cos_x, color="blue", linewidth=2.5)

# Plot 2:
sin_x = np.sin(x)

plt.subplot(1, 2, 2) #the figure has 1 row, 2 columns, and this plot is the second plot.
plt.plot(x, sin_x, color="red",  linewidth=2.5)

plt.show()
```

```python
# Plot 1:
x = np.arange(-2, 5, 0.1)
cos_x = np.cos(x)

plt.subplot(2, 1, 1)  # The figure has 2 row, 1 column, and this plot is the first plot.
plt.plot(x, cos_x, color="blue", linewidth=2.5)
plt.title("cosine function")

# Plot 2:
sin_x = np.sin(x)

plt.subplot(2, 1, 2) #the figure has 2 row, 1 column, and this plot is the second plot.
plt.plot(x, sin_x, color="red",  linewidth=2.5)
plt.title("sine function")

####You can add a title to the entire figure with the suptitle() function
plt.suptitle("My favourite functions")

# Adjust space between subplots
plt.subplots_adjust(hspace=0.5)  # Increase spacing
plt.show()
```

**Plotting one plot inside another in the same figure**

```python
import math
x = np.arange(0, math.pi*2, 0.05)
fig=plt.figure()  #create a figure

#axes1: The main plot is positioned at (10% from left, 10% from bottom, 80% width, 80% height).
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.55, 0.55, 0.3, 0.3]) # inset axes

y = np.sin(x)
axes1.plot(x, y, 'b')

axes2.plot(x, np.cos(x),'r')

axes1.set_title('sine')
axes2.set_title("cosine")

plt.show()
```

You can set your own color for each scatter plot with the color or the c argument

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, color = 'hotpink')

x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, color = '#48c699')

plt.show()
```

You can adjust the transparency of the dots with the alpha argument

You can change the size of the dots with the s argument

The Matplotlib module has a number of available colormaps.

A colormap is like a list of colors, where each color has a value that ranges from 0 to 100. You can include the colormap in the drawing by including the plt.colorbar() statement

```python
#Create random arrays with 100 values for x-points, y-points, colors and sizes

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

#The colormap (cmap) maps these numbers to colors.
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

#The colorbar (plt.colorbar()) shows how the numbers correspond to colors.
plt.colorbar()

plt.show()
```

**Pie chart**

With Pyplot, you can use the pie() function to draw pie charts. By default the plotting of the first wedge starts from the x-axis and moves counterclockwise
Image 01-04-2025 at 11.35.jpeg

```python
y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show()
```

Maybe you want one of the wedges to stand out? The explode parameter allows you to do that.

The explode parameter, if specified, and not None, must be an array with one value for each wedge.

Each value represents how far from the center each wedge is displayed

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.1, 0.1, 0.1, 0.1]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```

#Exercise 2
Create a figure with three subplots, aligned vertically. Create a bar graph for a and b on the first two, but at the last one draw both a and b, but put b on top of a! (Hint: use the bottom argument in the bar function) Make sure the color for a and b are consistent in all the subplots! Extra Challenge: Write the magnitude of each bar on top of it

```python
data = ['A', 'B', 'C']
a = np.array([3, 5, 2])
b = np.array([2, 6, 4])


# Create figure and subplots
fig = plt.figure(figsize = (10,15)) # 3 rows with one colum

#colors 


# Plot 1: Bar graph for 'a'

ax0 = fig.add_subplot(311)
ax0.bar(data, a, color='blue')
ax0.set_title("Bar Graph of 'a'")
## add the magnitude on each bar
for i, v in enumerate(a):
    ax0.text(i, v + 0.01, str(v), ha='center', fontsize=12)

# Plot 2: Bar graph for 'b'
ax1 = fig.add_subplot(312)
ax1.bar(data, b, color='green')
ax1.set_title("Bar Graph of 'b'")
## add the magnitude on each bar
for i, v in enumerate(b):
    ax1.text(i, v + 0.01, str(v), ha='center', fontsize=12)

# Plot 3: Stacked bar graph (b on top of a)
ax2 = fig.add_subplot(313)
ax2.bar(data, a, color='blue', label='a')
ax2.bar(data, b, bottom=a, color='green', label='b')  # b stacked on a
ax2.set_title("Stacked Bar Graph (b on top of a)")
## add the magnitude on each bar
for i, (va, vb) in enumerate(zip(a, b)):
    ax2.text(i, va + 0.1, str(va), ha='center', fontsize=12, color='white')  # a's value inside the bar
    ax2.text(i, va + vb + 0.01, str(vb), ha='center', fontsize=12, color='black')  # b's value inside b's bar

# Add legend to stacked plot
ax2.legend()


plt.show()
```
